```python
#Load Graph And Basic Info

G,mapping = snap.LoadEdgeListStr(snap.TUNGraph, "precedent-citation-actual-final-mod.txt", 0, 1,Mapping = True)
print("Number of Nodes: %d" % G.GetNodes())
print("Number of Edges: %d" % G.GetEdges())
print("Average Local Clustering Coeff: %f" % G.GetClustCf())
```

```python
Deg_Pairs = G.GetDegCnt()
Deg = []
Count_ = []

for item in Deg_Pairs:
    Deg.append(item.GetVal1())
    Count_.append(item.GetVal2())

plt.figure(figsize=(20,10))
plt.plot(Deg,Count_)
plt.grid()
plt.xlabel('Degree')
plt.ylabel('Count')
plt.title('Degree Distribution')
plt.savefig('Figures\DegreeDist.png')
```

```python
DegCenArr = np.array([])
dictionary = {}
for Node in G.Nodes():
    DegCenArr = np.append(DegCenArr, G.GetDegreeCentr(Node.GetId()))  
plt.plot(DegCenArr)
plt.grid()

#Top 5 nodes
Order = DegCenArr.argsort()[-5:][::-1]
print("Rank according to Degree Centrality :\n")
filehandler = open('Centrality/Map_Degree_Cent.txt', 'w+')
for i in Order:
    
    print("Doc Id :",mapping.GetKey(int(i)),"\tValue :",DegCenArr[i])
    dictionary[str(i)] = str(mapping.GetKey(int(i)))

filehandler = open('Centrality/Map_Degree_Cent.txt', 'w+')
filehandler.write(str(dictionary))
filehandler.close()
```

```python
ClosenessCenArr = np.array([])
dictionary = {}
for Node in G.Nodes():
    ClosenessCenArr = np.append(ClosenessCenArr, G.GetClosenessCentr(Node.GetId()))
    
plt.plot(ClosenessCenArr)
plt.grid()

#Top 5 nodes
Order = ClosenessCenArr.argsort()[-5:][::-1]

print("Rank according to Closeness Centrality :\n")
for i in Order:
    print("Doc Id :",mapping.GetKey(int(i)),"\tValue :",ClosenessCenArr[i])
    dictionary[str(i)] = str(mapping.GetKey(int(i)))

filehandler = open('Centrality/Map_Closeness_Cent.txt', 'w+')
filehandler.write(str(dictionary))
filehandler.close()
```

```python
BetCenArr = np.array([])
dictionary = {}
Nodes, Edges = G.GetBetweennessCentr(1.0)

for node in Nodes:
    BetCenArr = np.append(BetCenArr, Nodes[node])

plt.plot(BetCenArr)
plt.grid()

#Top 5 nodes
Order = BetCenArr.argsort()[-5:][::-1]

NumberOfNodes = G.GetNodes()
Normalize_factor = ((NumberOfNodes - 1)*(NumberOfNodes - 2))/2
print("Rank according to Betweenness Centrality :\n")
for i in Order:
    print("Doc Id :",mapping.GetKey(int(i)),"\tValue :",BetCenArr[i]/Normalize_factor)
    dictionary[str(i)] = str(mapping.GetKey(int(i)))

filehandler = open('Centrality/Map_Between_Cent.txt', 'w+')
filehandler.write(str(dictionary))
filehandler.close()
```

```python
EigenCenArr = np.array([])
dictionary = {}
Nodes = G.GetEigenVectorCentr()

for node in Nodes:
    EigenCenArr = np.append(EigenCenArr, Nodes[node])

plt.plot(EigenCenArr)
plt.grid()

#Top 5 nodes
Order = EigenCenArr.argsort()[-5:][::-1]

print("Rank according to Eigenvector Centrality :\n")
for i in Order:
    print("Doc Id :",mapping.GetKey(int(i)),"\tValue :",EigenCenArr[i])
    dictionary[str(i)] = str(mapping.GetKey(int(i)))

filehandler = open('Centrality/Map_Eigen_Cent.txt', 'w+')
filehandler.write(str(dictionary))
filehandler.close()
```

```python
# Create datasheet in the order ['docname',title','subjects','keywords']

subject_keys = open('subject_keywords.txt','r',encoding='utf8')
tokenizer = nltk.RegexpTokenizer(r"\w+")
lemmatizer = nltk.stem.WordNetLemmatizer()
doc_datasheet = []
title_datasheet = []
subject_datasheet = []
keywords_datasheet = []
for doc_ in subject_keys:
    b = doc_.split('-->')
    doc_datasheet.append(b[0])
    title_datasheet.append(b[1])
    sub_key = b[2]
    sub_key_arr = sub_key.split('$$$')
    
    
    temp = tokenizer.tokenize(sub_key_arr[0])                                #Preprocessing on subjects
    sub_preprocessed = []
    for subs in temp:
        sub_preprocessed.append(lemmatizer.lemmatize(subs.lower()))
    sub_preprocessed = np.array(sub_preprocessed)
    subject_datasheet.append(sub_preprocessed)
    
    keywords_datasheet.append(sub_key_arr[1].rstrip("\n"))
```

```python
# File titles acc to Centrality

filehandler = open('Centrality/Map_Degree_Cent.txt','r')
dictionary = ast.literal_eval(filehandler.read())
print("Rank according to Degree Centrality :\n")
for i in dictionary.values():
    title = title_datasheet[doc_datasheet.index(i)]
    print("Doc Id :",i,"\tTitle :",title)
filehandler.close()

filehandler = open('Centrality/Map_Closeness_Cent.txt','r')
dictionary = ast.literal_eval(filehandler.read())
print("\nRank according to Closeness Centrality :\n")
for i in dictionary.values():
    title = title_datasheet[doc_datasheet.index(i)]
    print("Doc Id :",i,"\tTitle :",title)
filehandler.close()

filehandler = open('Centrality/Map_Between_Cent.txt','r')
dictionary = ast.literal_eval(filehandler.read())
print("\nRank according to Betweenness Centrality :\n")
for i in dictionary.values():
    title = title_datasheet[doc_datasheet.index(i)]
    print("Doc Id :",i,"\tTitle :",title)
filehandler.close()

filehandler = open('Centrality/Map_Eigen_Cent.txt','r')
dictionary = ast.literal_eval(filehandler.read())
print("\nRank according to Eigenvector Centrality :\n")
for i in dictionary.values():
    title = title_datasheet[doc_datasheet.index(i)]
    print("Doc Id :",i,"\tTitle :",title)
filehandler.close()
```

```python
sub_evaluate = np.concatenate(subject_datasheet).ravel()

unique_words,count_unique = np.unique(sub_evaluate,return_counts = True)

plt.figure(figsize = (15,10))
plt.bar(unique_words,count_unique)
plt.xticks(rotation='vertical')
plt.grid()
plt.xlabel('Subjects',c='r')
plt.ylabel('Frequency',c='g')
plt.title('Frequency of each unique word in Subject',c='b')
plt.savefig('Figures\FreqCount.jpg')
```
